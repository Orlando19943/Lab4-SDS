# Recolecci√≥n de datos

```python
dir_path = "./MALWR/"
files = []
pe = []
```

```python
for path in os.listdir(dir_path):
    if os.path.isfile(os.path.join(dir_path, path)):
        files.append(path)
print(files)
```

```python
files.remove(".DS_Store")
```

## Desempaquetando los archivos

```python
for file in files:
    os.system("upx -d " + dir_path+file)
```

```python

for file in files:
    pe.append(pefile.PE(dir_path+file))
```

```python
df = pd.DataFrame(files, columns=["files"])
```

```python
# Creando las columnas de los dlls
dlls = []
for x in pe:
    for entry in x.DIRECTORY_ENTRY_IMPORT:
        dlls.append(entry.dll.decode("utf-8"))

dlls = list(set(dlls))

for dll in dlls:
    df[dll] = ""
```

```python
# Colocanco cada api en su columna del dll
apis = []
for i in range(len(pe)):
    for entry in pe[i].DIRECTORY_ENTRY_IMPORT:
        apis = []
        for imp in entry.imports:
            apis.append(imp.name.decode("utf-8"))
        df.at[i,entry.dll.decode("utf-8")]=apis
```

# Implementaci√≥n del modelo

```python
scaler = StandardScaler()
X = scaler.fit_transform(tf_idf)
```

## M√©todo del codo

```python
def new_kmeans(clusters: int):
    return KMeans(
        init = "random",
        n_clusters = clusters,
        n_init = 1,
        max_iter = 200,
        random_state = 12345
    )
```

```python
# function returns WSS score for k values from 1 to kmax
def show_elbow(points, kmax):
    # Inertia is calculated as the WSS if there are no weights (there aren't)
    # https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
    sse = [new_kmeans(x).fit(points).inertia_  for x in range(1, kmax+1)]    
    sns.lineplot(list(range(1, kmax+1)), sse, x="Cluster Amount", y="SSE")
```

```python
show_elbow(X, 10)
```

Basado en el m√©todo del codo, se determin√≥ que se deber√≠a de trabajar con aproximadamente de 4 cl√∫steres.

## Coeficiente de silueta

```python
from sklearn.metrics import silhouette_score

def show_silhouette(points, clusters):
    sil = []

    for k in range (2, clusters+1):
        model = new_kmeans(k).fit(points)
        labels = model.labels_
        sil.append(silhouette_score(points, labels))
    
    sns.lineplot(list(range(2, clusters+1)), sil)
```

```python
show_silhouette(X, 15)
```

A trav√©s del m√©todo de la silueta, se puede observar que 4 cl√∫steres representa un m√°ximo local. Esto, combinado con el m√©todo del codo demuestra que se deben usar 4 cl√∫steres.

## Coeficiente de jaccard

```python
###### Basado de: https://naps.com.mx/blog/coeficiente-de-jaccard-en-python-conjuntos/
def jaccard_similarity(set1, set2):
    intersection = set1.intersection(set2)
    union = set1.union(set2)
    return len(intersection) / len(union)
```

```python
colors = ["red", "yellow", "green", "cyan"]

labels = new_kmeans(4).fit(X).labels_
names = df['files'].tolist()
features = list(map(lambda x: set(x.split(', ')), df['api'].tolist()))
```

```python
def draw(graph):
    widths = list(nx.get_edge_attributes(graph,'weight').values())
    nx.draw(
        graph, 
        with_labels=True, 
        node_color=[colors[labels[i]] for i in graph.nodes],
        labels={ node: names[node] for node in graph.nodes },
        font_size=9,
        width=widths
    )
```

```python
graphs = [nx.Graph() for _ in range(4)]
treshold = 0.5

for cluster_index, graph in enumerate(graphs):
    cluster = [i for i, label in enumerate(labels) if label == cluster_index]
    other = [i for i in range(len(labels)) if i not in cluster]
    
    graph.add_nodes_from(cluster)
    
    for i, j in combinations(cluster, 2):
        jaccard = jaccard_similarity(features[i], features[j])
        if jaccard > treshold:
            graph.add_edge(i, j, weight=jaccard*3)
            
    for i, j in product(cluster, other):
        jaccard = jaccard_similarity(features[i], features[j])
        if jaccard > treshold:
            graph.add_edge(i, j, weight=jaccard*3)
```

Para evitar sobrecarga visual, se graficar√° cada cluster. Cada cluster incluir√° los elementos del cluster y cualquier otro elemento que tenga una similitud jaccard con un valor mayor a 0.5. 

El ancho de cada conexi√≥n entre nodos refleja el valor de jaccard. Cabe mencionar, que si dos nodos tuvieron un valor jaccard menor a 0.5, no se grafic√≥ la conexi√≥n entre ellos.
